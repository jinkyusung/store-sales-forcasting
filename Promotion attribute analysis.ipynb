```python
import pandas as pd
import collections


promo1 = pd.read_csv("dataset/promos/promo1.csv")
promo2 = pd.read_csv("dataset/promos/promo2.csv")
promo3 = pd.read_csv("dataset/promos/promo3.csv")
promo4 = pd.read_csv("dataset/promos/promo4.csv")
promo5 = pd.read_csv("dataset/promos/promo5.csv")


########## Data visualization ##########
promo1.plot(kind="scatter", x="Promotion1", y="Weekly_Sales", alpha=0.1)
promo2.plot(kind="scatter", x="Promotion2", y="Weekly_Sales", alpha=0.1)
promo3.plot(kind="scatter", x="Promotion3", y="Weekly_Sales", alpha=0.1)
promo4.plot(kind="scatter", x="Promotion4", y="Weekly_Sales", alpha=0.1)
promo5.plot(kind="scatter", x="Promotion5", y="Weekly_Sales", alpha=0.1)


########## Correlation Coefficient ##########
corr1 = promo1.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion1"]
corr2 = promo2.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion2"]
corr3 = promo3.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion3"]
corr4 = promo4.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion4"]
corr5 = promo5.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion5"]

print(f"Promo1 = {corr1}")
print(f"Promo2 = {corr2}")
print(f"Promo3 = {corr3}")
print(f"Promo4 = {corr4}")
print(f"Promo5 = {corr5}")
```
