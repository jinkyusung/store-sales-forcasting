# Analysis of `Store`
A `Weekly_Sales` of each store is quite related to **their average or median** of `Weekly_Sales` and its **trends**.  

Let's see 'Weekly_Sales` trends of each store.  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def modify_date(date):
    day = date[:2]
    month = date[3:5]
    year = date[6:]
    return year + month + day


train = pd.read_csv("dataset/train.csv")
train["Date"] = train["Date"].apply(modify_date)


plt.figure(figsize=(15, 30))
for store in range(1, 46):
    t = train[train["Store"] == store]
    ax = plt.subplot(9, 5, store)
    plt.title(f"store{store}") 
    ax.axes.xaxis.set_ticks([])
    ax.axes.yaxis.set_ticks([])
    plt.xlabel("Date")
    plt.ylabel("Weekly_Sales")
    plt.scatter(t["Date"], y=t["Weekly_Sales"])

    
```

```python
# Split train.csv
t1 = train[train["Store"] <= 5]
t2 = train[(5 < train["Store"]) & (train["Store"] <= 10)]
t3 = train[(10 < train["Store"]) & (train["Store"] <= 15)]
t4 = train[(15 < train["Store"]) & (train["Store"] <= 20)]
t5 = train[(20 < train["Store"]) & (train["Store"] <= 25)]
t6 = train[(25 < train["Store"]) & (train["Store"] <= 30)]
t7 = train[(30 < train["Store"]) & (train["Store"] <= 35)]
t8 = train[(35 < train["Store"]) & (train["Store"] <= 40)]
t9 = train[(40 < train["Store"]) & (train["Store"] <= 45)]


trains = [t1, t2, t3, t4, t5, t6, t7, t8, t9]

plt.figure(figsize=(15, 40))
for i, t in enumerate(trains, 1):
    plt.subplot(5, 2, i)
    
    sns.boxplot(x="Store", y="Weekly_Sales", hue="IsHoliday", data=t)    
```
