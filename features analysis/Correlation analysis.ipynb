```python
import pandas as pd


train = pd.read_csv("../dataset/train.csv")
train = train.fillna(0)

def get_day(date):
    return int(date[:2])

def get_month(date):
    return int(date[3:5]) 

def get_year(date):
    return int(date[6:])


train['Month'] = train['Date'].apply(get_month)
train['Day'] = train['Date'].apply(get_day)
train['Year'] = train['Date'].apply(get_year)

train
```

```python
########## Data visualization ##########
corr_test = train.copy()

# Store
corr_test.plot(kind="scatter", x="Store", y="Weekly_Sales", alpha=0.1)

# Date
corr_test.plot(kind="scatter", x="Date", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Day", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Month", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Year", y="Weekly_Sales", alpha=0.1)

# Social factor
corr_test.plot(kind="scatter", x="Temperature", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Fuel_Price", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Unemployment", y="Weekly_Sales", alpha=0.1)

# Promotion 1 ~ 5
corr_test.plot(kind="scatter", x="Promotion1", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Promotion2", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Promotion3", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Promotion4", y="Weekly_Sales", alpha=0.1)
corr_test.plot(kind="scatter", x="Promotion5", y="Weekly_Sales", alpha=0.1)
```

### Remark
    
**Store**: there exists that `Weekly_Sales` differences between each stores.  
We can group them by their each averge values of weekly sales.

|Group|Interval|
|:---:|:---:|
|A|1.5 < avg|
|B|1.0 <= avg < 1.5|
|C|avg < 1.0|

**Date(Day/Month/Year)**: Focus on the end of the __month(23 ~ 26)__ and the end of the __year(Nov, Dec)__.  
Inductively, it's because X-mas.

**Promotion 1~5**: See lines in every graph, `Promotion` value be zero.  
this happend due to our data preprocessing. Hence, we have to **except N/A values and re-analyze them**.
