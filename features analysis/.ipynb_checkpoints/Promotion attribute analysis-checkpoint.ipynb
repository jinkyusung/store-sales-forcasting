```python
import pandas as pd
import collections


promo1 = pd.read_csv("../dataset/promos/promo1.csv")
promo2 = pd.read_csv("../dataset/promos/promo2.csv")
promo3 = pd.read_csv("../dataset/promos/promo3.csv")
promo4 = pd.read_csv("../dataset/promos/promo4.csv")
promo5 = pd.read_csv("../dataset/promos/promo5.csv")


########## Data visualization ##########
promo1.plot(kind="scatter", x="Promotion1", y="Weekly_Sales", alpha=0.1)
promo2.plot(kind="scatter", x="Promotion2", y="Weekly_Sales", alpha=0.1)
promo3.plot(kind="scatter", x="Promotion3", y="Weekly_Sales", alpha=0.1)
promo4.plot(kind="scatter", x="Promotion4", y="Weekly_Sales", alpha=0.1)
promo5.plot(kind="scatter", x="Promotion5", y="Weekly_Sales", alpha=0.1)


########## Correlation Coefficient ##########
corr1 = promo1.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion1"]
corr2 = promo2.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion2"]
corr3 = promo3.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion3"]
corr4 = promo4.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion4"]
corr5 = promo5.corr()["Weekly_Sales"].round(2).sort_values(ascending=False)["Promotion5"]

print(f"Promo1 = {corr1}")
print(f"Promo2 = {corr2}")
print(f"Promo3 = {corr3}")
print(f"Promo4 = {corr4}")
print(f"Promo5 = {corr5}")
```

```python
def find_notnull(df, col):
    return df[df[col].notnull()]


def analyze(group):
    g1 = find_notnull(group, "Promotion1")
    g2 = find_notnull(group, "Promotion2")
    g3 = find_notnull(group, "Promotion3")
    g4 = find_notnull(group, "Promotion4")
    g5 = find_notnull(group, "Promotion5")
    
    g1.plot(kind="scatter", x="Promotion1", y="Weekly_Sales", alpha=0.1)
    g2.plot(kind="scatter", x="Promotion2", y="Weekly_Sales", alpha=0.1)
    g3.plot(kind="scatter", x="Promotion3", y="Weekly_Sales", alpha=0.1)
    g4.plot(kind="scatter", x="Promotion4", y="Weekly_Sales", alpha=0.1)
    g5.plot(kind="scatter", x="Promotion5", y="Weekly_Sales", alpha=0.1)
    
    corr1 = g1.corr()["Weekly_Sales"]["Promotion1"].round(2)
    corr2 = g2.corr()["Weekly_Sales"]["Promotion2"].round(2)
    corr3 = g3.corr()["Weekly_Sales"]["Promotion3"].round(2)
    corr4 = g4.corr()["Weekly_Sales"]["Promotion4"].round(2)
    corr5 = g5.corr()["Weekly_Sales"]["Promotion5"].round(2)

    print(corr1)
    print(corr2)
    print(corr3)
    print(corr4)
    print(corr5)
    print()
```

```python
groupA = pd.read_csv("dataset/groups/groupA.csv")
analyze(groupA)
```

```python
groupB = pd.read_csv("../dataset/groups/groupB.csv")
analyze(groupB)
```

```python
groupC = pd.read_csv("../dataset/groups/groupC.csv")
analyze(groupC)
```
